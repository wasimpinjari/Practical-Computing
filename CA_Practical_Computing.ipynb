```python
def rectangular_cuboid(L,B,H):
    A=[L,B,H]
    A.sort(reverse=True)
    print('area of one of the biggest faces of rectangular_cuboid is',A[0]*A[1],'dim=',A[0],'*',A[1])
    print('volume of a rectangular cuboid is',A[0]*A[1]*A[2],'dim=',A[0],'*',A[1],'*',A[2])
```

```python
rectangular_cuboid(30,15,20)
```

### Vectorised Matrix Multiplication

```python
import numpy as np
import time
def matrixmul(n):
    start = time.time()
    X=np.random.randint(1,5,(n,n))
    Y=np.random.randint(5,10,(n,n))
    Z=np.dot(X,Y)
    end = time.time()
    T=end-start
    #print(T)
    return Z ,T
    
```

```python
Time_1=[]
a=[10,30,50,100,500,1000]
for i in a:
    _,T= matrixmul(i)
    Time_1.append(T)
    
    
    print('new vaule---------------------',T)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(a,Time_1)
plt.xlabel('dimention of matrix')
plt.ylabel('Time')
plt.show()
```

### Non-Vectorised Matrix Multiplication

```python
def nonvector(n):
    A=np.random.randint(1,5,(n,n))
    B=np.random.randint(5,10,(n,n))
    result=np.zeros([n, n], dtype = int)
    start = time.time()
    for i in range(len(A)):

        # iterating by column by B
        for j in range(len(B[0])):

            # iterating by rows of B
            for k in range(len(B)):
                result[i][j] += A[i][k] * B[k][j]
    end = time.time()
    T=end-start
    #for r in result:
        #print(r)
    return T
```

```python
nonvector(10)
```

```python
Time_2=[]
a=[10,30,50,100,500,1000]
for i in a:
    T= nonvector(i)
    Time_2.append(T)
    
    
    print('new vaule---------------------',T)
```

```python
plt.plot(a,Time_2)
plt.xlabel('dimention of matrix')
plt.ylabel('Time')
plt.show()
```

```python
plt.plot(a,Time_2 , c='b')
plt.plot(a,Time_1, c='r')
plt.xlabel('dimention of matrix')
plt.ylabel('Time')
plt.show()
```
